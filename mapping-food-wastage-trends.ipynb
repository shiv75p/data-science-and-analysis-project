## Description

 Here's an explanation of the different components:

1. **Country**: This refers to the name of a specific country.
2. **Combined figures (kg/capita/year)**: This could represent the total amount of food consumed per capita in kilograms per year, which includes food consumed at households, retail, and food service establishments.
3. **Household estimate (kg/capita/year)**: This indicates the amount of food consumed per capita at households, measured in kilograms per year.
4. **Household estimate (tonnes/year)**: This is the total amount of food consumed by households in tonnes per year.
5. **Retail estimate (kg/capita/year)**: This represents the amount of food consumed per capita that was purchased from retail stores, measured in kilograms per year.
6. **Retail estimate (tonnes/year)**: The total amount of food consumed from retail sources in tonnes per year.
7. **Food service estimate (kg/capita/year)**: This refers to the amount of food consumed per capita at food service establishments (such as restaurants, cafes, etc.), measured in kilograms per year.
8. **Food service estimate (tonnes/year)**: The total amount of food consumed at food service establishments in tonnes per year.
9. **Confidence in estimate**: This could represent the reliability or accuracy of the provided estimates, possibly on a scale from low to high.
10. **M49 code**: This is a numeric code used to classify countries or regions for statistical and analytical purposes.
11. **Region**: The geographical area or region to which the country belongs.

## Import Libraries

```python
from seaborn import pairplot
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

## Scatter and Jointplot

```python
sns.jointplot(x='Household estimate (kg/capita/year)',y='Retail estimate (kg/capita/year)',data=df ,kind='reg')
plt.show()
```

```python
# Assuming 'df' is your DataFrame containing the data
sns.scatterplot(x='Household estimate (tonnes/year)', y='Retail estimate (tonnes/year)', data=df)
sns.regplot(x='Household estimate (tonnes/year)', y='Retail estimate (tonnes/year)', data=df, scatter=False)  # Adding the regression line
plt.show()
```

```python
sns.scatterplot(x='Household estimate (kg/capita/year)',y='Household estimate (tonnes/year)',data=df, hue='Region')
plt.show()
```

## Pie Plot

```python
#Confidence level of data
df['Confidence in estimate'].value_counts().plot(kind='pie',subplots=True)
plt.suptitle('Reliability of the data')
plt.ylabel("")
plt.show()
```

```python
#Region of data
df['Region'].value_counts().plot(kind='pie',subplots=True)
plt.suptitle('Reliability of the data')
plt.ylabel("")
plt.show()
```

## Ratio

```python
household_ratio = df['Household estimate (kg/capita/year)'] / df['combined figures (kg/capita/year)']
plt.hist(household_ratio, bins=50, color='green')
plt.xlabel('Household Ratio')
plt.ylabel('Frequency')
plt.title('Histogram of Household Ratio')
plt.show()
```

```python
household_ratio = df['Household estimate (tonnes/year)'] / df['Retail estimate (tonnes/year)']
plt.hist(household_ratio, bins=50, color='red')
plt.xlabel('Household Ratio')
plt.ylabel('Frequency')
plt.title('Histogram of Household Ratio')
plt.show()
```

```python
#Region Wise Mean Household estimate (tonnes/year)
plt.figure(figsize=(21,15))
average_wastage_household = df.groupby('Region')['Household estimate (tonnes/year)'].sum().sort_values(ascending=False).reset_index()
sns.barplot(data=average_wastage_household, y='Region', x='Household estimate (tonnes/year)')
plt.ylabel('Region')
plt.xlabel('Average Household estimate (tonnes/year)')
plt.title('Average Household estimate (tonnes/year) per Region')
plt.show()
```
